```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import gymnasium as gym
from stable_baselines3 import SAC
from IPython.display import HTML

env = gym.make("Pendulum-v1", render_mode="rgb_array")


N_steps = 500

frames = []
obs, _ = env.reset()
for _ in range(N_steps):
    action, _ = model.predict(obs, deterministic=True)
    obs, reward, done, truncated, info = env.step(action)
    frame = env.render()
    frames.append(frame)
    if done or truncated:
        obs, _ = env.reset()

env.close()

# Create matplotlib animation
fig = plt.figure()
img = plt.imshow(frames[0])

def animate(i):
    img.set_data(frames[i])
    return [img]

ani = animation.FuncAnimation(fig, animate, frames=len(frames), interval=30)

plt.close()
HTML(ani.to_html5_video())
```

```python
import gymnasium as gym
import numpy as np
import matplotlib.pyplot as plt

from stable_baselines3 import SAC
from stable_baselines3.common.callbacks import BaseCallback


class RewardLogger(BaseCallback):
    def __init__(self, verbose=0):
        super().__init__(verbose)
        self.episode_rewards = []
        self.episode_lengths = []
        self.episode_timesteps = []
        self._total_steps = 0

    def _on_step(self) -> bool:
        # Called at every environment step
        infos = self.locals.get("infos")
        dones = self.locals.get("dones")

        # For vectorized envs: len(dones) = n_envs
        if dones is None or infos is None:
            return True

        # Increase total step count by number of envs
        self._total_steps += len(dones)

        for i, done in enumerate(dones):
            if done and "episode" in infos[i]:
                ep_info = infos[i]["episode"]
                self.episode_rewards.append(ep_info["r"])
                self.episode_lengths.append(ep_info["l"])
                self.episode_timesteps.append(self._total_steps)

                if self.verbose > 0:
                    print(f"Episode finished: "
                          f"reward={ep_info['r']:.2f}, "
                          f"len={ep_info['l']}, "
                          f"timestep={self._total_steps}")

        return True
```

```python
# Create environment and wrap it so episode stats are added to info
env = gym.make("HalfCheetah-v4")
env = gym.wrappers.RecordEpisodeStatistics(env)

# Create SAC model
model = SAC("MlpPolicy", env, verbose=1)

# Train with callback
N = 400_000  # or whatever your assignment wants
reward_logger = RewardLogger(verbose=0)
model.learn(total_timesteps=N, callback=reward_logger)

env.close()
```

```python
plt.figure(figsize=(8, 5))
plt.plot(reward_logger.episode_timesteps,
         reward_logger.episode_rewards, marker="o", linestyle="-")
plt.xlabel("Timesteps")
plt.ylabel("Episode reward")
plt.title("SAC on HalfCheetah-v4: Learning Curve")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
settings = [0.3, 0.1, 0.03, 0.01]  # example values
loggers = {}

for ent in settings:
    env_exp = gym.make("HalfCheetah-v4")
    env_exp = gym.wrappers.RecordEpisodeStatistics(env_exp)

    logger = RewardLogger()
    model = SAC(
        "MlpPolicy",
        env_exp,
        ent_coef=ent,      # <--- key change
        verbose=0
    )
    model.learn(total_timesteps=N, callback=logger)
    env_exp.close()

    loggers[ent] = logger
```

```python
plt.figure(figsize=(8, 5))
for ent, logger in loggers.items():
    plt.plot(logger.episode_timesteps, logger.episode_rewards,
             label=f"ent_coef={ent}")
plt.xlabel("Timesteps")
plt.ylabel("Episode reward")
plt.title("Effect of ent_coef on SAC (HalfCheetah-v4)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
